```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('object.tiff')
sobel_x = cv2.Sobel(img,0,1,0,ksize=5)
sobel_y = cv2.Sobel(img,0,0,1,ksize=5)
sobel_xy = cv2.Sobel(img,0,1,1,ksize=5)
laplacian = cv2.Laplacian(img,0,ksize=5)
canny = cv2.Canny(img,20,200)
cv2.imshow('Original Image',img)
cv2.imshow('Sobel X',sobel_x)
cv2.imshow('Sobel y',sobel_y)
cv2.imshow('Sobel XY',sobel_xy)
cv2.imshow('Laplacian',laplacian)
cv2.imshow('Canny',canny)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread('baby.jpg')
img = cv2.resize(img,(800,500))
sobel_x = cv2.Sobel(img,0,1,0,ksize=5)
sobel_y = cv2.Sobel(img,0,0,1,ksize=5)
sobel_xy = cv2.Sobel(img,0,1,1,ksize=5)
laplacian = cv2.Laplacian(img,0,ksize=5)
canny = cv2.Canny(img,20,200)
cv2.imshow('Original Image',img)
cv2.imshow('Sobel X',sobel_x)
cv2.imshow('Sobel y',sobel_y)
cv2.imshow('Sobel XY',sobel_xy)
cv2.imshow('Laplacian',laplacian)
cv2.imshow('Canny',canny)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
