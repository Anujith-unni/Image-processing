```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_path = "baby.jpg" ## it should be  in the same folder
img=cv2.imread(img_path) ## reads the image as BGR
rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(rgb)
plt.show()
```

```python
img_path = "baby.jpg" ## it should be  in the same folder
img=cv2.imread(img_path) ## reads the image as BGR
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(gray,cmap='gray')
plt.show()
```

```python
img_path = "baby.jpg" ## it should be  in the same folder
img=cv2.imread(img_path)## reads the image as BGR
gray=cv2.imread(img_path,0)
plt.imshow(gray,cmap='gray')
plt.show()
```

```python
print(img.ndim)
print(rgb.ndim)
print(gray.ndim)
```

```python
print(img.shape)
print(rgb.shape)
print(gray.shape)
```

```python
print(img.dtype)
print(rgb.dtype)
print(gray.dtype)
```

```python
img_path = "baby.jpg"   ## it should be in the same folder
img = cv2.imread(img_path)
cv2.imshow("My Image",img)
cv2.waitKey(0)   ## wait till we press any key
cv2.destroyAllWindows()  ## close all windows related to opencv
```

```python
img_path = "baby.jpg"   ## it should be in the same folder
img = cv2.imread(img_path)
gray=cv2.imread(img_path,0)
cv2.imshow("RGB Image",img)
cv2.imshow("Grayscale",gray)
cv2.waitKey(0)   ## wait till we press any key
cv2.destroyAllWindows()  ## close all windows related to opencv
```
