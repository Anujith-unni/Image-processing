```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=np.zeros((512,512,3),np.uint8)+255
#img,pt1, pt2,color[,thickness]
cv2.line(img,(400,100),(100,400),(255,0,0),5)
cv2.line(img,(100,100),(400,400),(0,255,0),5)
cv2.line(img,(250,100),(250,400),(0,0,255),5)
plt.imshow(img)
plt.show()
```

## Draw Square/Rectangle


```python
img=np.zeros((512,512,3),np.uint8)+255
# img,pt1 ,pt2,color,thickness
cv2.rectangle(img,(400,100),(100,350),(255,250,0),5)
cv2.rectangle(img,(150,350),(350,200),(0,205,0),-1)
plt.imshow(img)
plt.show()
```

## Draw circle


```python
img=np.zeros((512,512,3),np.uint8)+255
# img,center ,radius,color,thickness
cv2.circle(img,(250,250),200,(255,0,0),5)
cv2.circle(img,(250,250),10,(0,0,255),-1)
plt.imshow(img)
plt.show()
```

## Draw Ellipse

```python
img=np.zeros((512,512,3),np.uint8)+255
# img,centre ,axes,angle,startangle,endangle,color,thickness
cv2.ellipse(img,(250,250),(200,100),0,0,360,(255,0,0),5)
plt.imshow(img)
plt.show()
```

##  Draw Polygon

```python
img=np.zeros((512,512,3),np.uint8)+255
points=np.array([[100,400],[100,200],[200,100],[300,200],[300,400],[200,500]])
#img, pts, isclosed,color,thickness
cv2.polylines(img,[points],True,(255,0,0),5)
plt.imshow(img)
plt.show()
```

```python
img=np.zeros((512,512,3),np.uint8)+255
points=np.array([[100,400],[100,200],[200,100],[300,200],[300,400],[200,500]])
#img, pts, isclosed,color,thickness
cv2.polylines(img,[points],False,(255,0,0),20)
plt.imshow(img)
plt.show()
```

```python
img=np.zeros((512,512,3),np.uint8)+255
points=np.array([[100,400],[100,200],[200,100],[300,200],[300,400],[200,500]])
#img, pts, isclosed,color,thickness
cv2.polylines(img,[points],False,(255,0,0),5)
plt.ylim([0,1000])
plt.imshow(img)
plt.show()
```

## write text

```python
img=np.zeros((512,512,3),np.uint8)+255
#img,text,org,fontFace,fontScale,color,thickness
cv2.putText(img,"Python",(120,250),cv2.FONT_HERSHEY_SCRIPT_COMPLEX,3,(255,0,0),5)
plt.grid()
plt.xlim([0,1000])
plt.ylim([1000,0])
plt.imshow(img)
plt.show()
```
