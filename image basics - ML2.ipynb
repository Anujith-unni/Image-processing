# create a blank grayscale image

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = np.zeros((20,20),np.uint32)
plt.imshow(img,cmap='gray')
plt.show()
```

```python
img=np.identity(20,np.uint32)*255 
plt.imshow(img,cmap='gray')
plt.show()
```

```python
img=np.identity(20,np.uint32)
plt.imshow(img,cmap='gray')
plt.show()
```

# colour image - rgb image

```python
img=np.zeros((20,20,3),np.uint32)
plt.imshow(img)
plt.show()
```

```python
img=np.zeros((20,20,3),np.uint32) + 255
plt.imshow(img)
plt.show()
```

```python
img=np.zeros((20,20,3),np.uint32)
img[:] = (0,255,0) ## rgb
plt.imshow(img)
plt.show()
```

```python
img=np.zeros((20,20,3),np.uint32)
img[:] = (35,149,34) ## rgb
plt.imshow(img)
plt.show()
```

```python
img=np.zeros((20,20,3)) + 0.5
plt.imshow(img)
plt.show()
```

```python
### region of interest (cropping of image)
img=np.zeros((20,20,3),np.uint32)
img[:] = (35,149,34) ## rgb
img[5:16,5:16] = (255,0,0)
plt.imshow(img)
plt.show()
```

```python
img=np.zeros((20,20,3),np.uint32)
img[:] = (35,149,34) ## rgb
for i in range(20):
    img[i,i] = (255,0,0)
plt.imshow(img)
plt.show()
```
